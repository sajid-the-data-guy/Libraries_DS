```python
#imoorting the matplotlib library
import matplotlib.pyplot as plt
```

```python
#instead of using print the show()function each time as well u execute this code will automatically execute the function 
%matplotlib inline 
```

```python
# importing numpy array
import numpy as np
#creating a numpy array
x = np.arange(1,20)
y = np.arange(21,40)
```

```python
x
```

```python
#ploting the x and y

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('scatter plot')
#saving the image in graph
plt.savefig('scatter.png')
```

```python
plt.scatter(x,y,c='g')

plt.plot(x,y)
```

```python
#creating a sub plot
plt.subplot(2,2,1)
plt.plot(x,y,c='r')
plt.subplot(2,2,2)
plt.plot(x,y,c='green')
plt.subplot(2,2,3)
plt.plot(x,y,c='blue')
plt.subplot(2,2,4)
plt.plot(x,y,'y*-')
```

```python
#sin wave diagram

pi = np.pi
x = np.arange(0, 4 * pi, 0.1)
y = np.sin(x)
plt.title("sine wave plot")

plt.plot(x,y)
```

```python
#bar plot
x =[1, 3, 5]
y =[7, 11, 3]

b1 = plt.bar(x,y)
```

```python
x2 = [2, 4 ,6]
y2 = [14, 2, 5]
plt.bar(x2,y2,color ='red')
```

```python
#combaining both plot
plt.bar(x,y)
plt.bar(x2,y2)
plt.title('Bar chart')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
#histogram
a = np.array([22, 32, 12, 45, 12, 53, 12, 22, 34, 12, 15, 35])
plt.hist(a,bins=10)
plt.title("histrogram")
plt.show()
```

```python
#box plot
data=[np.random.normal(0, std, 100)for std in range(1, 4)]
data
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
```

```python
#pie chart
labels = 'Maths', 'Statistic', 'English', 'Social','Science'
sizes = [215, 130, 245, 122, 300]
colors = ['green','red','yellow','pink','blue']
#plot
plt.pie(sizes, labels=labels, colors=colors)
```
